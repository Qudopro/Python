```python
import matplotlib.pyplot as plt         #Contains the Matplotlib library's graphing capabilities
```

```python
import numpy as np              #Include the function unique to summarize the dice rolls
```

```python
import random
```

```python
import seaborn as sns           #Contains the Seaborn library's graphing capabilities
```

```python
roll = [random.randrange(1,7) for i in range(60_000_000)]       #Create a list of 60,000,000 random dice values
```

```python
values, frequencies = np.unique(roll, return_counts=True)           #Determine the unique roll values
                                                            #return_counts=True count each unique value's number of occurrences
                                                            #return_counts=False only the list of unique values is returned
```

```python
title = f'Rolling a Six-Sided Die {len(roll):,} times'          #Bar plot's tile. It includes the number of dice rolls, and it displays the number with thousands separators with f-string
```

```python
sns.set_style('whitegrid')
```

```python
axes = sns.barplot(x=values, y=frequencies, palette='bright')       #Graph the dice frequencies using Seaborn's barplot function
axes.set_title(title)               #Display the title string centered above the plot
axes.set(xlabel='Die Value', ylabel='Frequency')            #Add labels to each axis (x, y)
axes.set_ylim(top=max(frequencies)*1.10)            #top change the maximum value represented by the y-axis; 10% taller than the tallest bar
for bar, frequency in zip(axes.patches, frequencies):       #patches contains two-dimensional colored shapes that represent the plot's bars
    text_x = bar.get_x()+bar.get_width()/2.0            #Center x-coordinate where the text will appear: LEFT-EDGE + HALF OF THE BAR'S WIDTH
    text_y = bar.get_height()                           #Y-Coordinate where the text will appear. get_height represents the bar's top
    text = f'{frequency:,}\n{frequency/len(roll):.3%}'
    axes.text(text_x, text_y, text, ha='center', va='bottom')   #Display the text above the bar.
    # First two arguments specify the text's x-y position.
    # ha = horizontal alignment (center around the x-coordinate),
    # va = vertical alignment (the bottom of the text with at the y-coordinate)


```

```python
plt.show()
```

```python
plt.cla()           #Clear the existing graph
```
