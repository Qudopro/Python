```python
numbers = [10, 3, 29, 19, 84, 2, 1, 5]
```

```python
numbers.sort()
print(id(numbers))
print(numbers)
```

```python
numbers.sort(reverse=True)
print(id(numbers))
print(numbers)
```

```python
ascending_numbers = sorted(numbers)
print(id(sorted(numbers)))
print(ascending_numbers)
```
