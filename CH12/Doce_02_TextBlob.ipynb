```python
#GETTING DEFINITIONS, SYNONYMS AND ANTONYMS FROM WORDNET
```

```python
happy = Word('happy')
```

```python
happy.definitions
```

```python
happy.define()
```

```python
happy.synsets           #originalWord'sLemmatizedForm, a = adjective, n = noun, v = verb, r = adverb, s = adjective satellite, 01 = index number of the corresponding meaning in the WordNet database
```

```python
synonyms = set()
```

```python
for synset in happy.synsets:
    for lemma in synset.lemmas():                   #returns the synonymous word as a string
        synonyms.add(lemma.name())
```

```python
synonyms
```

```python
lemmas = happy.synsets[0].lemmas()
```

```python
lemmas
```

```python
lemmas[0].antonyms()
```
