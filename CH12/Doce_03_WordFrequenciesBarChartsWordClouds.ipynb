```python
from pathlib import Path
```

```python
from textblob import TextBlob
```

```python
text = Path('pg1513.txt').read_text(encoding="UTF-8", errors="ignore")
```

```python
blob = TextBlob(Path('pg1513.txt').read_text(encoding="UTF-8", errors="ignore"))
```

```python
from nltk.corpus import stopwords
```

```python
stop = stopwords.words('english')
```

```python
items = blob.word_counts.items()            #Tuple word and frequency
```

```python
items = [item for item in items if item[0] not in stop]         #Obtenemos la palabra de cada tupla y la eliminamos si es un stopwords
```

```python
from operator import itemgetter
```

```python
sorted_items = sorted(items, key=itemgetter(1), reverse=True)       #Ordenamos las tuplas por el valor de la frecuencia (reverse = True -> descending order)
```

```python
top20 = sorted_items[1:21]              #Only get 20 elements, ignoring element 0 (')
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(top20, columns=['word', 'count'])
```

```python
df
```

```python
axes = df.plot.bar(x='word', y='count', legend=False)       #axis x -> word column, and y -> count column
```
