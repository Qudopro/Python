```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
document = nlp('In 1994, Tim Berners-Lee founded the World Wide Web Consortium (W3C), devoted to developing web technologies')
```

```python
for entity in document.ents:            #Returns a tuple of Span objects representing the named entities found in the Doc
    print(f'{entity.text}: {entity.label_}')
```
