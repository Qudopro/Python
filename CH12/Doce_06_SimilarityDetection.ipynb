```python
from codecs import ignore_errors

import spacy
```

```python
nlp = spacy.load('en_core_web_sm')              #English model
```

```python
from pathlib import Path
```

```python
document1 = nlp(Path('pg1513.txt').read_text(encoding='utf-8', errors="ignore"))
```

```python
document2 = nlp(Path('pg20288.txt').read_text(encoding='utf-8', errors="ignore"))
```

```python
document1.similarity(document2)         #Get a value from 0.0 (not similar) to 1.0 (identical) indicating how similar the documents are
```

```python
document3 = nlp('In 1994, Tim Berners-Lee founded the World Wide Web Consortium (W3C), devoted to developing web technologies')
```

```python
document1.similarity(document3)
```

```python
document2.similarity(document3)
```
