```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
#To perform dimensionality reductions
```

```python
from sklearn.manifold import TSNE           #TSNE estimator to perform dimensionality reductions
```

```python
tsne = TSNE(n_components=2, random_state=11)            #Reducing a dataset's features to two dimensions with n_components
```

```python
#To reduction... training the estimator (fit method) and use the estimator to transform the data into the specified number of dimensions (transform method=
```

```python
reduced_data = tsne.fit_transform(digits.data)      #Train and transform the data. Every call trains the estimator
```

```python
reduced_data.shape          #Data has only 2 features and these could be quite different from the dataset's original features
```

```python
#Visualizing the Reduced Data
```

```python
import matplotlib.pyplot as plt
```

```python
dots = plt.scatter(
    reduced_data[:,0],          #Column 0       as x axes
    reduced_data[:,1],          #Column 1       as y axes
    c='black'                   #color of dots
)
```

```python
#Visualizing the Reduced Data with different colors for each digit
```

```python
dots = plt.scatter(
    reduced_data[:,0],
    reduced_data[:,1],
    c=digits.target,                            #Specifies that the target values determine the dot color
    cmap=plt.cm.get_cmap('nipy_spectral_r',10)  #Specifies a color map to use when coloring the dots, 10 distinct colors
)
colorbar = plt.colorbar(dots)                   #Adds a color bar key to knwo which digit each color represents
```
