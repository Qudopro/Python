```python
{1,3,5} < {7,3,5,1}             #All elements in the left operand are in the right operand, and sets are not equal. Proper subset
```

```python
{1,2,3} < {4,5,6,7}
```

```python
{1,3} <= {3,5,1}                #All elements in the left operand are in the right operand, and sets might be equal. Improper subset
```

```python
{1,3}.issubset({3,5,1})         #Improper subset
```

```python
{1,2}.issubset({3,5,1})
```

```python
{1,3,5} > {3,5,1}               #All elements in the right operand are in the left operand, and the left operand has more elements. Proper Superset
```

```python
{1,3,5,7} > {3,5,1}
```

```python
{1,3} >= {3,1}                #All elements in the right operand are in the left operand, and the sets might be equal. Improper superset
```

```python
{1,3,5}.issuperset({3,5,1})     #Improper superset
```

```python
{1,3}|{5,6,9}               #Union operator. All unique elements
```

```python
{1,3}.union([3,5,1])        #Union operation
```

```python
{1,3,5} & {6,3,0}           #Intersection operator. All unique elements that the two sets have in common
```

```python
{1,3,5}.intersection([3,5,1])   #Intersection operation
```

```python
{1,3,5} - {6,3,0}               #Difference operator. Elements in the left operand that are not in the right operand
```

```python
{1,3,5,7}.difference([3,5,7])       #Difference operation
```

```python
{1,3,5} ^  {2,3,4}          #Symmetric difference operator. Elements of both sets that are not in common with one another
```

```python
{1,3,5,7}.symmetric_difference([3,5,7])     #Symmetric difference operation.
```

```python
{1,3,5}.isdisjoint({3,5,7})             #Disjoint operation. Don't have any common elements
```

```python
numbers = {1,3,5}
numbers |= {3,5,7}
print(numbers)
```

```python
numbers.update(range(10))           #It performs a union operation
print(numbers)
```

```python
# &= -> intersection_update()
# -= -> difference_update()
# ^= -> symmetric_difference_update()
```

```python
numbers.add(20) #Insert its argument in the set, if the argument is not already in the set
```
