```python
from tweetutilities import get_API
```

```python
api = get_API()
```

```python
tweets = []         #List to store the tweets we collect
```

```python
counts = {"total_tweets": 0, "locations":0}          #Number of tweets collected and tweets with locations
```

```python
from locationlistener import LocationListener
```

```python
location_listener = LocationListener(api, counts_dict=counts, tweets_list = tweets, topic = 'football', limit = 50)
```

```python
#Search only english tweets
```

```python
import tweepy
```

```python
stream = tweepy.Stream(auth=api.auth, listener=location_listener)
```

```python
stream.filter(track=['football'], languages=['en'], is_async=False)
```

```python
#Cleaning the data
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(tweets)
```

```python
df = df.dropna()                #Remove any row with NaN
```

```python
#Plot the tweets locations
```

```python
import folium
```

```python
usmap = folium.Map(location=[39.8283, -98.5795], tiles='Stamen Terrain', zoom_start=5, detect_retina=True)

#latitude and longitude coordinates for the map's center point
#zoom start indicates the map's initial zoom level
#detect_retina enables folium to detect high resolution screens
```

```python
for t in df.itertuples():           #Create tuples from each row of the Dataframe
    text = ': '.join([t.screen_name, t.text])
    popup = folium.Popup(text, parse_all=True)          #Display the text
    marker = folium.Marker((t.latitude, t.longitud), popup = popup)     #Create object Marker using a tuple to specify the latitude and longitude
    marker.add_to(usmap)            #To specify the Map that will display the Marker
```

```python
usmap.save('tweet_map.html')            #save the map in a HTML file
```
