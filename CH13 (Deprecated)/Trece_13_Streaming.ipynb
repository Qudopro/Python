```python
import tweepy
```

```python
import keys
```

```python
#Authenticate with Twitter and create a Tweepy API Object
auth = tweepy.OAuthHandler(keys.consumer_key, keys.consumer_secret)
auth.set_access_token(keys.access_token, keys.access_token_secret)
api = tweepy.API(auth, wait_on_rate_limit=True)
```

```python
from tweetlistener import TweetListener
```

```python
tweet_listenter = TweetListener(api)
```

```python
tweet_stream = tweepy.Stream(auth=api.auth, listener=tweet_listenter)       #Manages the connection to the Twitter stream creating a Tweepy Stream object
```

```python
#Begins the streaming process
tweet_stream.filter(track=['Mars Rover'], is_async=True)

#track parameter pass a list of search terms
#Mode asynchronous tweet stream, allows your code to continue executing while your listener waits to receive tweets, so you can continueinteract with other parts of the application
```
