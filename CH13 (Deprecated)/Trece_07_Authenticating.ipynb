```python
#Get an Account's followers with pagination (via Cursor)
```

```python
followers = []              #List to store the tweepy.models.User objects
```

```python
cursor = tweepy.Cursor(api.followers, screen_name='nasa')           #for NASA's account. api.followers = name of the method to call, screen_name is passed to api.followers method
```

```python
for account in cursor.items(10):                #Get 10 followers in each pagination
    followers.append(account.screen_name)
```

```python
print('Followers:',' '.join(sorted(followers, key=lambda s: s.lower())))        #Second argument to sorted: determine how the elements of followers are sorted
```

```python
#Twitter Trends API
```

```python
trends_available = api.trends_available()           #Get a list of all locations for which Twitter has trending topics
```

```python
#Worldwide trending topics
```

```python
world_trends = api.trends_place(id=1)
```

```python
trends_list = world_trends[0]['trends']         #One element list containing a dictionary, which trends is key
```

```python
trends_list = [t for t in trends_list if t['tweet_volume']]         #Return only trends with more than 10,000 tweets
```

```python
from operator import itemgetter
```

```python
trends_list.sort(key=itemgetter('tweet_volume'), reverse=True)          #Sorted trends in descending order by tweet_volume
```

```python
for trend in trends_list[:5]:               #Display the names of the top five trending topics
    print(trend['name'])
```

```python
#Create a Word Cloud from Trending Topics
```

```python
topics = {}                 #Create a dictionary of key-value pairs consisting of the trending topic names and tweet_volumes
```

```python
for trend in trend_list:
    topics[trend['name']] = trend['tweet_volume']
```

```python
from wordcloud import WordCloud                     #Use to create word cloud
```

```python
wordcloud = WordCloud(width=1600, height=1600, prefer_horizontal=0.5, min_font_size=10, colormap='prism', background_color='white')
#prefer_horizontal = 0.5 suggests that 50% of the words should be horizontal
```

```python
wordcloud = wordcloud.fit_words(topics)
```

```python
wordcloud = wordcloud.to_file('mask_circle.png)
```
