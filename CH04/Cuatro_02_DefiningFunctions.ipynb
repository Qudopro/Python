```python
def maximum(x, y, z):
    """Return the maximum of three values."""
    max_value = x
    if max_value < y:
        max_value = y
    if max_value < z:
        max_value = z
    return max_value
```

```python
maximum(1, 2, 3)
```

```python
maximum(2.2, 3.3, 1.1)
```

```python
maximum("Hola", "Crayola", "Saludos")
```

```python
maximum(13.5, -7, 20)
```

```python
max([14,25,46,76])
```
