```python
def maximum(value1, value2, value3):
    """Return the maximum of three values"""
    max_value = value1
    if max_value < value2:
        max_value = value2
    if max_value < value3:
        max_value = value3
    return max_value
```

```python
maximum(12, 26, 37)
```

```python
maximum(12.3, 45.6, 9.7)
```

```python
maximum('yellow', 'red', 'orange')
```

```python
maximum(12, 43.5, -3)
```
