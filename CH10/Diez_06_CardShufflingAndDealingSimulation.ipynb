```python
from deck import DeckOfCards
deck_of_cards = DeckOfCards()
```

```python
from pathlib import Path
```

```python
path = Path('.').joinpath('card_images')                #Create a Path object for the current folder + subfolder card_images
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
figure, axes_list = plt.subplots(nrows=4, ncols=13)              #4 rows and 13 columns
```

```python
for axes in axes_list.ravel():                          #One dimensional view of a multidimensional array
    axes.get_xaxis().set_visible(False)
    axes.get_yaxis().set_visible(False)                 #Hide  the x- and y-axes
    image_name = deck_of_cards.deal_card().image_name           #deals a card and gets its image_name
    img = mpimg.imread(str(path.joinpath(image_name).resolve()))                    #Añade image_name al Path y resolve determina la ruta completa de la imagen. Luego se carga la imagen con imread
    axes.imshow(img)                            #Muestra la imagen actual en el subplot
```

```python
figure.tight_layout()
```
