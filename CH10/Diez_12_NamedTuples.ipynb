```python
from collections import namedtuple
```

```python
Card = namedtuple('Card', ['face', 'suit'])   #(new type's name, list of strings representing the identifiers)
```

```python
card = Card(face='Ace', suit='Spades')
```

```python
card.face
```

```python
card.suit
```

```python
card
```

```python
values = ['Queen', 'Hearts']
```

```python
card = Card._make(values)
```

```python
card
```

```python
card._asdict()
```
