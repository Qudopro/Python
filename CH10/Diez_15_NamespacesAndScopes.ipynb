```python
z = 'global z'
```

```python
def print_variables():
    y = 'local y in print_variables'            #Local scope
    print(y)                                    #Built-in scope (first looks local, then global, then built-in)
    print(z)
```

```python
print_variables()
```

```python
y
```

```python
z
```

```python
#Si accedes a un identificador desde una funcion anidada, primero buscará en el ambito de la función anidada, luego en la función envolvente, luego en el global y por ultimo en el built-in (LEGB)
```
