```python
'Valid' if re.fullmatch(r'\d{5}', '02215') else 'Not Valid'     # / begins each of the predefined character classes
```

```python
'Valid' if re.fullmatch(r'\d{5}', '0215') else 'Not Valid'          # /d -> represents a digit. {5} -> Quantifier. Repeats 5 times \d
```

```python
#\d -> any digit.
#\D -> Any character that is not a digit
#\s Any whitespace character
#\S any character that is not a whitespace character
#\w any word character
#\W any character that is not a word character
```

```python
'Valid' if re.fullmatch('[A-Z][a-z]*', 'Wally') else 'Not Valid'        # *  matches zero or more occurrences of the subexpression to its left
```

```python
'Match' if re.fullmatch('[*+$]', '*') else 'Not Match'
```

```python
'Valid' if re.fullmatch('labell?ed', 'labelled') else 'Not Valid'       # ? matches 0 or one occurrences of a subexpression
```

```python
result = re.search('Python', 'Python is fun')
result.group() if result else 'not found'
```

```python
result = re.search('fun!', 'Python is fun')
result.group() if result else 'not found'
```

```python
result = re.match('Python', 'Python is fun')
result.group() if result else 'not found'
```

```python
result = re.search('Python', 'python is fun', flags=re.IGNORECASE)
result.group() if result else 'not found'
```

```python
result = re.search('^Python', 'python is fun')      #The expression matches only the beginning of a string
result.group() if result else 'not found'
```

```python
result = re.search('^fun', 'python is fun')
result.group() if result else 'not found'
```

```python
result = re.search('Python$', 'python is fun')      # $ at the end of a regular expression, the expression matches only the end of a string
result.group() if result else 'not found'
```

```python
result = re.search('fun$', 'python is fun')      # $ at the end of a regular expression, the expression matches only the end of a string
result.group() if result else 'not found'
```

```python
contact = 'Wally White, Home: 555-555-5432, Work: 555-555-5555'
```

```python
re.findall(r'\d{3}-\d{3}-\d{4}', contact)
```

```python
for phone in re.finditer(r'\d{3}-\d{3}-\d{4}', contact):
    print(phone.group())
```

```python
text = 'Charlie Cyan, email: demo@deitel.com'
```

```python
pattern = r'([A-Z][a-z]+ [A-Z][a-z]+), email: (\w+@\w+\.\w{3})'     # () capture substrings in a match. '.' match one character
```

```python
result = re.search(pattern, text)
result.groups()
```

```python
result.group(1)
```
