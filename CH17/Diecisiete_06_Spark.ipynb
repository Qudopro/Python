```python
import nltk
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')
```

```python
#Configuring a SparkContext
```

```python
#Configuration Options by a SparkConf object
```

```python
from pyspark import SparkConf, SparkContext
```

```python
configuration = SparkConf() .setAppName('RomeoAndJulietCounter').setMaster('local[*]')
       #Specify the Spark application's name
       #Specify the Spark cluster's URL (in this case, on your local computer). * Indicates that Spark should run our code using the same number of threads as there are cores on the computer
```

```python
#Creating the SparkContext
```

```python
from pyspark import SparkContext
```

```python
sc = SparkContext(conf=configuration)
```

```python
from textblob.utils import strip_punc
```

```python
tokenized = sc.textFile('RomeoAndJulietCounter.txt')\
    .map(lambda line: strip_punc(line, all=True).lower())\
    .flatMap(lambda line: line.split())

#Loads the lines of text from txt and returns it as an RDD of strings that represent each line
#Remove all punctuation with TextBlob's strip_punch function and to convert each line of text to loweracse. Returns a new RDD
#Map each line of text into its words and produces a single list of words. Retrun a new RDD representing all the words in Romeo and Juliet
```

```python
#Filter to create a new RDD with no stop words remaining
```

```python
filtered = tokenized.filer(lambda word: word not in stop_words)
```

```python
from operator import add
```

```python
word_counts = filtered.map(lambda word: (word, 1)).reduceByKey(add)
#Add is a function that tells method reduceByKey to add the counts for tuples that contain the same word (the key)
```

```python
#Filter the RDD to keep only those words with 60 or more occurrences
```

```python
filtered_counts = word_counts.filter(lambda item: item[1]>=60)
```

```python
#Sorting in descending order the list of tuples by their counts, using collect to obtain the results
```

```python
from operator import itemgetter
```

```python
sorted_items = sorted(filtered_counts.collect(), key=itemgetter(1), reverse=True)
```

```python
#Display the result in table format
```

```python
max_length = max([len(word) for word, count in sorted_items])
```

```python
for word, count in sorted_items:
    print(f'{word:>{max_length}}: {count}')
```
