```python
def linear_search(data, search_key):
    for index, value in enumerate(data):
        if value == search_key:
            return index
    return -1
```

```python
import numpy as np
```

```python
np.random.seed(11)
```

```python
values = np.random.randint(10, 91, 10)
```
