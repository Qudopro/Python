```python
def factorial(number):
    """Return factorial of number"""
    #Determines whether the terminating condicion number is True (base case)
    if number <= 1:
        return 1

    return number * factorial(number - 1)               #Recursive call -> number * (number-1)! until number! = 1
```

```python
for i in range(11):
    print(f'Factorial of {i}: {factorial(i)}')
```
